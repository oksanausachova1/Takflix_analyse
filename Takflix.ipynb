#**Merge multiple dataframes into one, sorts the rows by the 'Created' column, creates a new column with datetime values, and checks if all months in a certain range are present in the resulting dataframe**

````python
# Sort the merged dataframe by the 'Created' column
merged_df.sort_values(by=['Created'], inplace=True)

# Convert the 'Created' column to a datetime format and create a new column 'datetime'
merged_df['datetime'] = pd.to_datetime(merged_df['Created'], format='%Y-%m-%d %H:%M')

# Create a new column 'year_month' with the year and month from the 'datetime' column
merged_df['year_month'] = merged_df['datetime'].dt.strftime('%Y-%m')

# Get the first and last row of the dataframe
first_row = merged_df.iloc[0]
last_row = merged_df.iloc[-1]

# Convert the 'year_month' values from the first and last row to datetime format
start = pd.to_datetime(first_row['year_month'], format='%Y-%m')
end = pd.to_datetime(last_row['year_month'], format='%Y-%m')

# Create a new list of all the months between the start and end date (inclusive) in the '%Y-%m' format
all_dates = pd.date_range(start, end, freq='MS').strftime('%Y-%m')

# Get a list of unique year_month values in the dataframe
unique_dates = merged_df['year_month'].unique()

# Check if all the months between the start and end date are present in the dataframe
flag = False
for date in all_dates:
  if date not in unique_dates:
    print(f"Місяць {date} відсутній у датафреймі.")
    flag = True

# Print a message indicating whether all months are present or not
if not flag:
  print("Всі місяці присутні")

# Print the merged dataframe
merged_df.head(10)

````

#**Sorts the merged DataFrame by datetime, groups it by owner's email and year-month, sums the values, pivots the resulting DataFrame, fills the missing values with 0, and calculates the cumulative sum along the columns**

````python
# Sort the merged_df by datetime
merged_df = merged_df.sort_values('datetime')

# Group the dataframe by Owner's Email and year_month and calculate the sum of each group
grouped_df = merged_df.groupby(["Owner's Email", "year_month"]).sum()

# Pivot the table so that the year_month values become columns and the By Purchase values become cell values
# Fill any missing values with 0 and calculate cumulative sum along the columns
pivot_df = grouped_df.reset_index().pivot(index="Owner's Email", columns="year_month", values="By Purchase").fillna(0).cumsum(axis=1)
````

````python
pivot_df.head(300)
````

#**Calculates the loyalty coefficients for each month based on the pivot dataframe and visualizes the result with a bar chart**

````python
# Calculate loyalty coefficients for each month
loyalty_coeffs = pivot_df.apply(lambda x: round(sum(x > 1) / sum(x > 0) * 100), axis=0)
````

````python

# Create a DataFrame for the loyalty coefficients
loyalty_df = pd.DataFrame({'Month': loyalty_coeffs.index, 'Loyalty Coefficient': loyalty_coeffs.values})

# Create a bar chart using Plotly
fig = px.bar(loyalty_df, x='Month', y='Loyalty Coefficient', labels={'Loyalty Coefficient': 'Coefficient of Loyalty'}, title='Coefficient of Loyalty in Month')
fig.update_layout(xaxis_tickangle=-45)

# Show the plot
fig.show()
````

#**Calculates the gifts coefficient for each month based on the total number of gifts given compared to the total number of purchases made and then plots it in a bar chart**

````python
# Group data by year_month and sum the values
grouped_df = merged_df.groupby(["year_month"]).sum()

# Calculate the gifts coefficient as the ratio of gifts to purchases
grouped_df['gifts_coef'] = round(grouped_df['By Gift'] / grouped_df['By Purchase'] *100)

# Reset the index to make "year_month" a column
grouped_df = grouped_df.reset_index()

# Create a bar chart using Plotly
fig = px.bar(grouped_df, x='year_month', y='gifts_coef', labels={'gifts_coef': 'Gifts Coefficient'}, title='Gifts Coefficients by Month')
fig.update_layout(xaxis_tickangle=-45)

# Show the plot
fig.show()

````

#**Creates masks to identify specific promo codes and uses them to update the "By Purchase" column of the "merged_df" DataFrame**

````python
# Extract the 'Promo code' column from the merged dataframe
promo = merged_df['Promo code']

# Create boolean masks for each type of promo code
mask_subc = promo.str.startswith('SUBC')
mask_8mm = promo.str.startswith('8mm')
mask_16mm = promo.str.startswith('16mm')
mask_sub = promo.str.startswith('SUB')

# Combine the boolean masks with an OR operator to create a single mask for all relevant promo codes
mask_any = mask_subc | mask_8mm | mask_16mm | mask_sub

# Use the combined mask to set the 'By Purchase' column to 1 for all rows with relevant promo codes
merged_df.loc[mask_any, 'By Purchase'] = 1
````

#**Create a dictionary called email_to_movies that maps each email address in the loyal_clients_list DataFrame to a list of movies associated with that email**

````python
# create an empty dictionary to map email addresses to movies
email_to_movies = {}

# iterate over each row in the loyal_clients_list DataFrame
for index, row in loyal_clients_list.iterrows():
    # extract the film name and email address from the current row
    film = row['Film']
    email = row["Owner's Email"]

    # if the current email address already has movies associated with it, append the new movie to the list
    if email in email_to_movies:
        email_to_movies[email].append(film)
    # if the current email address does not have any movies associated with it yet, create a new list with the current movie
    else:
        email_to_movies[email] = [film]

# print the final email_to_movies dictionary
print(email_to_movies)
````

#**Let us find top 3 films first watching from loyal clients**

````python
# create an empty dictionary to map movies to view counts
movies_to_views = {}

# iterate over each email address in the email_to_movies dictionary
for email in email_to_movies:
    # get the first movie associated with the current email address
    movie = email_to_movies[email][0]

    # if the current movie has not been seen before, set its view count to 1
    if movie not in movies_to_views:
        movies_to_views[movie] = 1
    # if the current movie has been seen before, increment its view count by 1
    else:
        movies_to_views[movie] += 1

# sort the movies by view count in descending order and store them in a list of tuples
sorted_movies = sorted(movies_to_views.items(), key=lambda x: x[1], reverse=True)
# calculate the total number of views for all movies
total_views = 0
for film in sorted_movies:
    total_views += film[1]

# print the total number of views for all movies
print(total_views)
# print the list of movies and their view counts in descending order
sorted_movies

````
